```python
#Checking for missing values

df = pd.read_csv(r'G:\DataSets\movies.csv')

nan = df.isnull()

print(nan)
```

```python
#Dropping rows with missing values

df.dropna()
```

```python
#Finding data types for columns

df.dtypes
```

```python
# Filling na with zero value 

df = df.fillna(0)
```

```python
# Changing data type for budget and gross column

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
df
```

```python
#Sorting data on gross column 

df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
#Drop all duplicates

df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earings')

plt.ylabel('Budget for Film')

plt.show
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Finding correlation 
# Methods i use are ( pearson, kendall, spearman )

df.corr(method='pearson')
```

```python
# Visualizing correlation matrix using seaborn

matrix = df.corr(method='pearson')

sns.heatmap(matrix, annot=True)

plt.title('Correlation Matric for Numeric columns')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Finding correlation pairs

correlation_mat = df.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Columns with high correlation

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
